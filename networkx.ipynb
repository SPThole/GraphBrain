```python
#initiating the graph
g = nx.Graph()
g.add_node(0,Name='A',salary="1K",dept="DS")
g.add_node(1,Name='B',salary ="1K",dept="DS")
g.add_node(3,Name='C',salary = "5K",dept="DS")
g.add_node(4,Name='D',salary = "10K",dept="DE")
print(f"Number of nodes in graph g {g.number_of_nodes()}")

#Bertter way
g.add_nodes_from([(5,{"Name":'E',"salary":"10K","dept":"DS"}),(6,{"Name":'F',"salary":"20K","dept":"DS"})])
print(f"Number of nodes in graph g {g.number_of_nodes()}")

print("Name",nx.get_node_attributes(g,"Name"))
print("salary",nx.get_node_attributes(g,"salary"))
```

```python
# Adding edges
g.add_edge(6,5)
g.add_edge(6,4)
g.add_edge(5,3)
g.add_edge(3,0)
g.add_edge(3,1)

print(g.edges)
print(g.nodes)
```

```python
nx.draw(g,labels=nx.get_node_attributes(g,"Name"))
```

```python
# adjacent nodes to the given node
g.adj[6]
```

```python
DG = nx.DiGraph(name="EXAMPLE")
DG.add_edge(2, 1,name='A')
DG.add_edge(1, 3,name='B')
DG.add_edge(2, 4,name='C')
DG.add_edge(1, 2,name='D')
```

```python
print("edges",DG.nodes)
print("nodes",DG.edges)
print("Graph attr",DG.name)
print("Getting the attributes",nx.get_edge_attributes(DG,"name"))
# Connected to 2
print("Connected to 2",DG.degree[2])
print("Degree at once",DG.degree([1,2]))

# accessing the edge like array
print("accessing the edge",DG[1][3]["name"])
nx.draw(DG)

# removing the nodes
DG.add_node("A")
DG.remove_node("A")

print(DG.nodes)

print("For Loop operations",DG.edges.data('name'))

#successors while degree reports the sum of in_degree and out_degree even though that may feel inconsistent at times.
print(list(DG.successors(1)))
print(DG.in_degree(2))
print(DG.out_degree(2))

#neighbors
print(list(DG.neighbors(2)))
```
